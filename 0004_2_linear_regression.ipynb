```python
import math
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Latex
```

```python
np.random.seed(1337)
observations = np.array([
    [1971, 7.74],
    [1974, 8.51],
    [1978, 10.27],
    [1982, 11.80],
    [1985, 12.52]
])
X = np.array([observations[:,0]])
y = np.array([observations[:,1]]).T
X = np.vstack((np.ones_like(X), X)).T
```

<font size="5">
\begin{align}
\Theta = (X^T X)^{-1} X^T y
\end{align}
</font>

```python
one = np.linalg.pinv(np.dot(X.T, X))
two = np.dot(X.T, y)
theta = np.dot(one, two)
```

<font size="5">
    \begin{align}
    \hat{y} = h_\Theta(x) = \Theta x
    \end{align}
</font>

```python
y_hat = np.dot(X, theta)
new_val = np.array([[1990, 1995]])
new_val = np.vstack((np.ones_like(new_val), new_val)).T
y_more_hat = np.dot(new_val, theta)
```

```python
fig = plt.figure(num=None, figsize=[14,14])
fig.patch.set_facecolor("#FF0000")
fig.patch.set_alpha(0)

ax = fig.add_subplot(111)
ax.grid(True, which="both", color="#2f5372", linestyle="-", alpha=1)
ax.patch.set_alpha(0)

ax.plot(observations[:,0], observations[:,1], 'o')
ax.plot(observations[:,0], y_hat)
ax.plot(new_val[:,1], y_more_hat)

# plt.show()
```
